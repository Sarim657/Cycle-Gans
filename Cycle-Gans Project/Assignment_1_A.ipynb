Q4 Write a Pandas function to find the count of unique species.

```python
df = pd.read_csv(url, header=None)

df.rename(columns={0:'sepallength', 1:'sepalwidth', 2:'petallength', 3:'petalwidth', 4:'species'}, inplace=True)

uniqspecies = list()
for i in range(0, len(df['species'])):
    if df['species'][i] in uniqspecies:
        continue
    else:
        uniqspecies.append(df['species'][i])

print(len(uniqspecies))
```

# Q5 Write a Pandas function to bin the petal length (3rd) column of iris_2d to form a text array, such that if petal length is:
- Less than 3 –> ‘small’
- 3-5 –> ‘medium’
- greater than equal to 5 –> ‘large’

```python
for i in range(0, len(df['petallength'])):
    if df['petallength'][i] < 3:
        df['petallength'][i] = 'small'
    elif df['petallength'][i] >= 3 and df['petallength'][i] < 5:
        df['petallength'][i] = 'medium'
    else:
        df['petallength'][i] = 'large'

print(df)
```
