```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import models,layers
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
path='/content/drive/MyDrive/Train.csv'
df=pd.read_csv(path)
df.head()
```

```python
label_encoder = LabelEncoder()
df['MonthOfMatch'] = label_encoder.fit_transform(df['MonthOfMatch'])
df['Team1_Venue'] = label_encoder.fit_transform(df['Team1_Venue'])
df['Team2_Venue'] = label_encoder.fit_transform(df['Team2_Venue'])
df['Team1_Innings'] = label_encoder.fit_transform(df['Team1_Innings'])
df['Team2_Innings'] = label_encoder.fit_transform(df['Team2_Innings'])
```

```python
df.head()
print(df.shape)
```

```python
'''from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df[['Team1','Team2','Stadium','HostCountry','Team1_Venue','Team2_Venue','Team1_Innings','Team2_Innings','MonthOfMatch','MatchWinner']] = scaler.fit_transform(df[['Team1','Team2','Stadium','HostCountry','Team1_Venue','Team2_Venue','Team1_Innings','Team2_Innings','MonthOfMatch','MatchWinner']])'''
```

```python
x_train=df[['Team1','Team2','Stadium','HostCountry','Team1_Venue','Team2_Venue','Team1_Innings','Team2_Innings','MonthOfMatch']]
y_train=df['MatchWinner']
```

```python
x_test=x_train.iloc[2100:]
y_test=y_train.iloc[2100:]
print(x_test)
print('==================================================================================================================================================')
x_train=x_train.iloc[:2100]
y_train=y_train.iloc[:2100]
print(x_train)
```

```python
model=models.Sequential()

model.add(layers.Dense(2056,input_shape=(9,),activation='relu'))
model.add(layers.Dropout(0.2))

model.add(layers.Dense(1028,activation='relu'))
model.add(layers.Dropout(0.2))

model.add(layers.Dense(512,activation='relu'))
model.add(layers.Dropout(0.2))

model.add(layers.Dense(16,activation='softmax'))

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

model.fit(x_train,y_train,epochs=1000)
```

```python
model.evaluate(x_test,y_test)
```

```python
from sklearn.metrics import classification_report
y_pred=model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)
print(classification_report(y_test, y_pred_classes))
```
